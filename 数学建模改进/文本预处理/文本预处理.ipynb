```python
import os
import seaborn as sns
import pandas as pd
import matplotlib.pyplot as plt
from itertools import chain
import jieba.posseg as pseg
from wordcloud import WordCloud
#设置显示风格
plt.style.use('fivethirtyeight')
plt.rcParams['font.sans-serif']=['SimHei']
```

```python

train_tang_path=r'E:\math\train\唐诗'
train_song_path=r'E:\math\train\宋诗'
length=[]
labels=[]
for i in [train_tang_path,train_song_path]:
    labels.append(len(os.listdir(i)))
    for title in os.listdir(i):
        poet_path=os.path.join(i,title)
        with open(poet_path,'r',encoding='utf-8') as fp:
            #print(len(fp.read()))
            length.append(len(fp.read()))
```

```python
sns.countplot(x=labels)
# 主要关注count长度分布的纵坐标, 不需要绘制横坐标, 横坐标范围通过dist图进行查看
plt.xticks([1,0])
plt.title('训练集')
plt.show()
```

```python
test_tang_path = r'E:\math\test\唐诗'
test_song_path = r'E:\math\test\宋诗'
length_ = []
labels_ = []
for i in [test_tang_path, test_song_path]:
    labels_.append(len(os.listdir(i)))
    for title in os.listdir(i):
        poet_path = os.path.join(i, title)
        with open(poet_path, 'r', encoding='utf-8') as fp:
            #print(len(fp.read()))
            length_.append(len(fp.read()))
```

```python
sns.countplot(x=labels_)
# 主要关注count长度分布的纵坐标, 不需要绘制横坐标, 横坐标范围通过dist图进行查看
plt.xticks([1, 0])
plt.title('测试集')
plt.show()
```

```python
sns.countplot(x=length)
# 主要关注count长度分布的纵坐标, 不需要绘制横坐标, 横坐标范围通过dist图进行查看
plt.xticks([])
plt.title('训练集')
plt.show()
```

```python
sns.countplot(x=length_)
# 主要关注count长度分布的纵坐标, 不需要绘制横坐标, 横坐标范围通过dist图进行查看
plt.xticks([])
plt.title('测试集')
plt.show()
```
